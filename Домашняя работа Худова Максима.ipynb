```python
import numpy as np # linear algebra 
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
# stats models: regression fitting
import statsmodels.formula.api as smf
# data visualization
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
# Задача 2.1.1
sns.regplot(data=df, x='sleep', y='totwrk', ci=None, line_kws={"color": "r"})
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='sleep~totwrk', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='sleep~-1+totwrk', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
# Задача 2.1.2
sns.regplot(data=df, x='sleep', y='age', ci=None, line_kws={"color": "r"})
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='sleep~-1+age', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
# Задача 2.1.3
sns.regplot(data=df, x='sleep', y='totwrk', ci=None, line_kws={"color": "r"})
```

```python
# подгонка прямой
fitted_polynom = smf.ols(formula='sleep~totwrk+I(totwrk**2)', data=df).fit()
# коэффициенты с округлением
fitted_polynom.params.round(2)
```

```python
# Задача 2.2.1
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/Labour.csv')
df
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='output~capital', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
sns.regplot(data=df, x='output', y='capital', ci=None, line_kws={"color": "r"})
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='output~-1+capital', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
sns.regplot(x=np.log(df['capital']), y=np.log(df['output']), ci=None, line_kws={"color": "r"})
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='np.log(output)~np.log(capital)', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
# подгонка прямой
fitted_line = smf.ols(formula='np.log(output)~-1+np.log(capital)', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
# Задача 2.2.2
sns.regplot(data=df, x='output', y='labour', ci=None, line_kws={"color": "r"})
# подгонка прямой
fitted_line = smf.ols(formula='output~labour', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
sns.regplot(data=df, x='labour', y='output', ci=None, line_kws={"color": "r"})
# подгонка прямой
fitted_line = smf.ols(formula='output~-1+labour', data=df).fit()
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
sns.regplot(x=np.log(df['output']), y=np.log(df['labour']), ci=None, order=2, line_kws={"color": "r"})
# подгонка прямой
fitted_line = smf.ols(formula='np.log(output)~np.log(labour)', data=df).fit() 
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
sns.regplot(x=np.log(df['output']), y=np.log(df['labour']), ci=None, order=2, line_kws={"color": "r"})
# подгонка прямой
fitted_line = smf.ols(formula='np.log(output)~-1+np.log(labour)', data=df).fit() 
# коэффициенты с округлением
fitted_line.params.round(2)
```

```python
# Задача 2.2.3
sns.regplot(x=np.log(df['output']), order=2, y=np.log(df['capital']), ci=None, line_kws={"color": "r"})
```

```python
fitted_polynom = smf.ols(formula='np.log(output)~np.log(capital)+I(np.log(capital)**2)', data=df).fit()
fitted_polynom.params.round(2)
```

```python
#Задача 2.2.4
sns.regplot(x=np.log(df['output']), order=2, y=np.log(df['labour']), ci=None, line_kws={"color": "r"})
```
